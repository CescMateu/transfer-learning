```python
import os

data_path = '/gpfs/projects/bsc28/optretina/data/'
illnesses = ['altpig', 'dmae', 'excavation', 'membrana', 'nevus']
#illnesses = ['excavation']
file_prefix = 'anonymized_2*'
destination_dir = 'data/retina_data_subset'

#if not os.path.isdir(destination_dir):
#    print('Destination dir does not exist, creating one at {}'.format(destination_dir))
#    os.system('mkdir {}'.format(destination_dir))
```

```python
error = False
for illness in illnesses:
    ill_dir = 'u_{}_symbolic_512'.format(illness)
    illness_complete_dir = os.path.join(destination_dir, ill_dir)
    if not os.path.isdir(illness_complete_dir):
        print(illness_complete_dir)
    os.system('mkdir {}'.format(illness_complete_dir))
    
    for mode in ['test', 'train', 'validation']:
        mode_complete_dir = os.path.join(destination_dir, ill_dir, mode)
        os.system('mkdir {}'.format(mode_complete_dir))
        
        for class_type in ['normal', illness]:
            if class_type == 'excavation':
                class_type = 'excavacion' # Mispelling when the directories where created
            print('Retrieving files for class: {}'.format(class_type))
            class_type_complete_dir = os.path.join(destination_dir, ill_dir, mode, class_type)
            os.system('mkdir {}'.format(class_type_complete_dir))
            
            # Join the path to the files
            file_wildcard = os.path.join(data_path, ill_dir, mode, class_type, file_prefix)
            # Create the destination path
            destination = os.path.join(destination_dir, ill_dir, mode, class_type)
            
            print('Dowloading files from {} to {}'.format(file_wildcard, destination))
            # Execute the scp command for each one of the previous directories
            command = 'scp dt01:{} {}'.format(file_wildcard, destination)
            print(command)
            s = os.system(command)
            
            if s != 0:
                print('Downloading interrumped with status: {}'.format(s))
                error = True
            else:
                print('Download finished')
            print('============================')
            
if error:
    print('Script finished with some errors!')
else:
    print('Script finished without errors!')
```
